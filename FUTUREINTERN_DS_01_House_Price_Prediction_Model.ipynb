```python
# training data
HP_train = pd.read_csv('HP_train.csv')
HP_test= pd.read_csv('HP_test.csv')
```

```python
# Display basic info
HP_train.info()
HP_test.head()
```

```python
# Drop the 'Id' column as it's not a feature, and separate the target variable 'SalePrice'
X = HP_train.drop(columns=['SalePrice', 'Id'])
y = HP_train['SalePrice']
```

```python
# Identify categorical and numerical columns
categorical_cols = [col for col in X.columns if X[col].dtype == "object"]
numerical_cols = [col for col in X.columns if X[col].dtype in ['int64', 'float64']]
```

```python
# Define transformers for numerical and categorical data
numerical_transformer = SimpleImputer(strategy='mean')
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
# Prepare the test data (drop 'Id' for prediction)
X_test = HP_test.drop(columns=['Id'])
```

```python
# Make predictions
from sklearn.pipeline import Pipeline
test_preds = pipeline.predict(X_test)
```

```python
# Prepare the submission DataFrame
submission = pd.DataFrame({'Id': HP_test['Id'], 'SalePrice': test_preds})
```

```python
# Save to CSV
submission.to_csv('house_price_predictions.csv', index=False)
print("Predictions saved to house_price_predictions.csv")
```

```python
# Define the model
model = RandomForestRegressor(random_state=0)
```

```python
# Create a pipeline that combines preprocessing and the model
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('model', model)
                          ])
```

```python
# Split data into training and validation subsets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)

```

```python
# Train the model
pipeline.fit(X_train, y_train)
```

```python
# Predict on validation data
y_pred = pipeline.predict(X_valid)
```

```python
# Evaluate the model
mae = mean_absolute_error(y_valid, y_pred)
rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
```

```python
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```
